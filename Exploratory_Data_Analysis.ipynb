```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic_data_updated.csv')
df
```

### 3. Feature Screening

```python
# check the null value
df.isnull().sum()
```

```python
# checking duplicate value
df.duplicated().sum()

# remove dublicate value
df.drop_duplicates(inplace=True)
```

### 4. Statistical Profiling

```python
df.info()
```

```python
df.describe()
```

### 5. Univariate Analysis( Categorical Value Analysis -1 )

```python
sns.countplot(data=df, x=df['Survived'])
```

```python
# generate percentige
(df['Survived'].value_counts() / len(df)) * 100
```

```python
Survive_count = df['Survived'].value_counts()
Survive_count
survived_label = df['Survived'].unique()
survived_label

plt.pie(Survive_count, labels = survived_label, autopct="%1.1f")
plt.title('Pie chart of survival rate')
plt.show()
```

### 7.Univariate Analysis ( Numerical Column Analysis - 1)

```python
sns.histplot(df['Age'],bins=50)

plt.title('Histogram of Age')
plt.xlabel('Ages')
plt.ylabel('frequency')

plt.show()
```

```python
sns.kdeplot(df['Age'])
plt.title('KDE plot of Age')
plt.xlabel('Ages')
plt.ylabel('probabilty distribution')

plt.show()
```

## Multivariate Analysis

### 2.2 & 2.3 Multivariate Analysis- 1 ( Countplot with Hue )

```python
sns.countplot(x=df['Sex'], hue=df['Survived'])
```

```python
gender_split = df.groupby("Sex")['Survived'].value_counts(normalize=True)
gender_split
```

```python
sns.countplot(x=df['Pclass'],hue=df['Survived'])
```

```python
pclass_split = df.groupby('Pclass')['Survived'].value_counts(normalize=True)

print(pclass_split)
```

# Inheritance

```python
class Phone:
  def __init__(self, name, model, bettery):
    self.name = name
    self.model = model
    self.bettery = bettery

  def chager (self, persen):
    return f"Phone is {self.name}, now charge is {persen}"


class SmartPhone(Phone):
  def __init__(self, name, model, bettery, company):

    super().__init__(name, model, bettery)
    self.company = company

  def relase (self, date):
    return f"The phone {self.name} relase date is {date}"


# call the function
phone = Phone("Nokia", "Fz", 100)
print(phone.chager(90))
# inhertance
phone02 = SmartPhone("Iphone", "13ProMax", 100, "apple")
print(phone02.relase("Jan 27"))
```

## Polymorphsim


```python
class Smart_Camera :
  def __init__(self, name, model):
    self.name = name
    self.model = model

  def ModelUpgrade(self, data):
    return f"The {self.name} upgrate is {data}"

class DSLR(Smart_Camera):
  def __init__(self, name, model):
    super().__init__(name, model)
  def ModelUpgrade (self, data):
    return f"Sorry you can not update {self.name}, this featers: {data}"

camera = Smart_Camera("Smart Camera", "phone")
print(camera.ModelUpgrade("Sencor"))
dslr = DSLR("DSLR", "New")
print(dslr.ModelUpgrade("Iq"))
```
